```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score, roc_auc_score, roc_curve, auc
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, ExtraTreesClassifier, AdaBoostClassifier, GradientBoostingClassifier
from xgboost import XGBClassifier
from catboost import CatBoostClassifier

# Mount Google Drive if using Colab
from google.colab import drive
drive.mount('/content/drive')

# Read the data
veriler = pd.read_excel('/content/drive/My Drive/Colab Notebooks/Fibroid_Multi_Slice_Analysis/T2_mutli_slice/T2_All/2.xlsx')

# Calculate correlation matrix
correlation_matrix = veriler.corr()


# Plot correlation matrix
plt.figure(figsize=(12, 10))  # Increase the figure size
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", annot_kws={"fontsize": 8})  # Increase font size of annotations
plt.title('Correlation Matrix', fontweight='bold')
plt.savefig('/content/drive/My Drive/Colab Notebooks/Fibroid_Multi_Slice_Analysis/T2_mutli_slice/T2_All/ROC_Curves.png', dpi=300, bbox_inches='tight')
plt.show()



# Separate the features (X) and target variable (Y)
X = veriler.iloc[:, 0:29]
Y = veriler.iloc[:, 29:30]

# Perform one-hot encoding on the target variable
Y = pd.get_dummies(Y, drop_first=True)


# Split the data into training and test sets
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.33, random_state=42)

# Normalize the features
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)

# Define the classifiers
classifiers = [
    LogisticRegression(random_state=0),
    SVC(kernel='rbf', random_state=0, probability=True),
    DecisionTreeClassifier(random_state=0),
    RandomForestClassifier(random_state=0),
    ExtraTreesClassifier(random_state=0),
    XGBClassifier(),
]

# Define the labels for the classifiers
classifier_labels = [
    'Logistic Regression', 'Support Vector Machine','Decision Tree ', 'Random Forest ', 'Extra Tree ',
     'XGBOOST Classifier'
]

# Perform predictions and calculate ROC AUC scores for each classifier
plt.figure(figsize=(10, 8))

for clf, label in zip(classifiers, classifier_labels):
    clf.fit(x_train, y_train)
    y_pred = clf.predict(x_test)  # Predict probabilities of positive class
    fpr, tpr, _ = roc_curve(y_test, y_pred)
    roc_auc = auc(fpr, tpr)
    plt.plot(fpr, tpr, label=f'{label} (AUC = {roc_auc:.2f})')

# Set plot properties
plt.plot([0, 1], [0, 1], 'k--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate', fontweight='bold')
plt.ylabel('True Positive Rate', fontweight='bold')
plt.title('Receiver Operating Characteristic', fontweight='bold')
plt.legend(loc='lower right')

# Save the ROC curve plot
plt.savefig('/content/drive/My Drive/Colab Notebooks/Fibroid_Multi_Slice_Analysis/T2_mutli_slice/T2_All/ROC_Curves.png', dpi=300, bbox_inches='tight')

# Display the ROC curve plot
plt.show()

# Perform predictions and print classification reports for each classifier


class_labels = ['NPV<90%', 'NPV>90%']

for clf, label in zip(classifiers, classifier_labels):
    clf.fit(x_train, y_train)
    y_pred = clf.predict(x_test)
    
    print(f'------------------ {label} ------------------')
    cm = confusion_matrix(y_test, y_pred)
    print('Confusion Matrix:\n', cm)
    
    # Plot confusion matrix
    plt.figure(figsize=(6, 4))
    sns.heatmap(cm, annot=True, cmap='Blues', fmt='d')
    plt.title(f'Confusion Matrix - {label}', fontweight='bold')
    plt.xlabel('Predicted Values', fontweight='bold')
    plt.ylabel('Actual Values ', fontweight='bold')
    plt.xticks([0.5, 1.5], class_labels)
    plt.yticks([0.5, 1.5], class_labels)
    plt.savefig(f'/content/drive/My Drive/Colab Notebooks/Fibroid_Multi_Slice_Analysis/T2_mutli_slice/T2_All/{label}_Confusion_Matrix.png', dpi=300, bbox_inches='tight')
    plt.show()
    print('Classification Report:\n', classification_report(y_test, y_pred, target_names=class_labels))
    print('Accuracy:', accuracy_score(y_test, y_pred))
    print('AUC Score:', roc_auc_score(y_test, y_pred))
    print()
```

```python
import pandas as pd
from tabulate import tabulate
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score, roc_auc_score, roc_curve, auc, recall_score, precision_score
from sklearn.metrics import f1_score

# Initialize empty lists to store the performance metrics
precision_scores = []
recall_scores = []
accuracy_scores = []
f1_scores = []

# Iterate over the classifiers and calculate the performance metrics
for clf, label in zip(classifiers, classifier_labels):
    clf.fit(x_train, y_train)
    y_pred = clf.predict(x_test)
    
    # Calculate performance metrics
    precision = precision_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    accuracy = accuracy_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)
    
    # Append the metrics to the respective lists
    precision_scores.append(precision)
    recall_scores.append(recall)
    accuracy_scores.append(accuracy)
    f1_scores.append(f1)

# Create a DataFrame to store the performance metrics
performance_df = pd.DataFrame({
    'Model': classifier_labels,
    'Precision': precision_scores,
    'Recall': recall_scores,
    'Accuracy': accuracy_scores,
    'F1-Score': f1_scores
})

# Set the index of the DataFrame as the 'Model' column
performance_df.set_index('Model', inplace=True)

# Display the performance metrics table
print(tabulate(performance_df, headers='keys', tablefmt='fancy_grid', colalign=('center', 'center', 'center', 'center')))
```
